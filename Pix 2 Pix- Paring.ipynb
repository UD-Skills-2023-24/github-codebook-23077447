```python
import os
import numpy as np
import cv2
from PIL import Image

# Image size
target_width = 512
target_height = 256

# Number of images
num = 2

# Dataset A path
pathA = 'D:/BARTLETT/Skill module/a'

# Dataset B path
pathB = 'D:/BARTLETT/Skill module/b'

# Save path
save_path = 'D:/BARTLETT/Skill module/c'

# Load images
def load_imgs(path, num, target_width, target_height):
    imgs = []
    for i in range(num):
        file_name = f"{i}.jpg"
        img_path = os.path.join(path, file_name)
        img = Image.open(img_path)
        
        # Resize while preserving aspect ratio
        img = img.resize((target_width, target_height), Image.ANTIALIAS)
        
        imgs.append(img)
    return imgs

# Convert PIL image to OpenCV format
def pil_to_cv(pil_img):
    return cv2.cvtColor(np.array(pil_img), cv2.COLOR_RGB2BGR)

# Pair images, save and preview each pair using OpenCV
for count in range(num):
    img_pair = Image.new('RGB', (target_width * 2, target_height))
    img_pair.paste(imgA[count], (0, 0))
    img_pair.paste(imgB[count], (target_width, 0))

    # Convert PIL image to OpenCV format
    img_pair_cv = pil_to_cv(img_pair)

    # Save the paired image
    paired_image_path = os.path.join(save_path, f"{count}.jpg")
    img_pair.save(paired_image_path)

    # Preview the paired image using OpenCV
    cv2.imshow("Paired Image", img_pair_cv)
    cv2.waitKey(500)  # Display each image for 500 ms
    cv2.destroyAllWindows()

    print(f"\npair {count + 1} completed")

print(f"{num} pairs ready for pix2pix!")
```

```python
#for maintaing the size ratio

import os
import numpy as np
import cv2
from PIL import Image

# Image size
target_width = 512
target_height = 256

# Number of images
num = 156

# Dataset A path
pathA = 'D:/BARTLETT/Skill module/a'

# Dataset B path
pathB = 'D:/BARTLETT/Skill module/b'

# Save path
save_path = 'D:/BARTLETT/Skill module/c'

# Load images
def load_imgs(path, num, target_width, target_height):
    imgs = []
    for i in range(num):
        file_name = f"{i}.jpg"
        img_path = os.path.join(path, file_name)
        img = Image.open(img_path)
        
        # Resize while maintaining aspect ratio
        img.thumbnail((target_width, target_height))
        background = Image.new('RGB', (target_width, target_height), (255, 255, 255))
        background.paste(img, (int((target_width - img.size[0]) / 2), int((target_height - img.size[1]) / 2)))
        imgs.append(background)
    return imgs

# Convert PIL image to OpenCV format
def pil_to_cv(pil_img):
    return cv2.cvtColor(np.array(pil_img), cv2.COLOR_RGB2BGR)

# Pair images, save and preview each pair using OpenCV
imgA = load_imgs(pathA, num, target_width, target_height)
imgB = load_imgs(pathB, num, target_width, target_height)

for count in range(num):
    img_pair = Image.new('RGB', (target_width * 2, target_height))
    img_pair.paste(imgA[count], (0, 0))
    img_pair.paste(imgB[count], (target_width, 0))

    # Convert PIL image to OpenCV format
    img_pair_cv = pil_to_cv(img_pair)

    # Save the paired image
    paired_image_path = os.path.join(save_path, f"{count}.jpg")
    img_pair.save(paired_image_path)

    # Preview the paired image using OpenCV
    cv2.imshow("Paired Image", img_pair_cv)
    cv2.waitKey(500)  # Display each image for 500 ms
    cv2.destroyAllWindows()

    print(f"\npair {count + 1} completed")

print(f"{num} pairs ready for pix2pix!")
```
